### Research Question 1 :1- what trend of movies based on runtimes, popularity and revenues over years?

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.
df_TMDb.groupby('release_year')['runtime'].mean().plot(kind = 'line', figsize = (15, 15)) 
plt.title('Average runtime of movies over years', fontsize = 16) 
plt.xlabel('release_year', fontsize = 16)
plt.ylabel('runtime', fontsize = 16);
```

```python
df_TMDb.groupby('release_year')['popularity'].mean().plot(kind = 'line', figsize = (15, 15)) 
plt.title('Average popularities of movies over years', fontsize = 16) 
plt.xlabel('release_year', fontsize = 16)
plt.ylabel('Popularity', fontsize = 16);
```

```python
df_TMDb.groupby('release_year')['revenue'].mean().plot(kind = 'line', figsize = (15, 15)) 
plt.title('Average revenue of movies over years', fontsize = 16) 
plt.xlabel('release_year', fontsize = 16)
plt.ylabel('revenue', fontsize = 16);
```

### Research Question 2  (what is the highest revenue of movies over years?)

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.
df_TMDb.sort_values('revenue', ascending = False).head(1)
```

### Research Question 3 ( what is favorite movies over years?)

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.
df_TMDb.sort_values('vote_average', ascending = False).head(1)
```

### Research Question 4 (what is count of movies for each year?)

```python
df_TMDb.groupby('release_year')['release_year'].count().plot(kind = 'bar', figsize = (15, 15)) 
plt.title('number of movies produced over years', fontsize = 16) 
plt.xlabel('release year', fontsize = 16)
plt.ylabel('Numbs of Movies', fontsize = 16);
```
